```python
yx = lambda x: x**2
ye = lambda x: np.exp(x)

fig, (sp1, sp2) = plt.subplots(nrows = 1, ncols= 2, figsize= (5,5))
sp1.plot(x, yx(x), color= 'green', lw= 3,  label= 'x**2')
sp2.plot(x, ye(x), color= 'red', lw= 3, label= 'e**x')

# setting the scale
plt.xscale('linear') # normal
plt.yscale('log') # logarithmic

sp1.legend()
sp2.legend()
plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.

```python
by_month = fitbit.groupby('Months').agg({'Steps': 'median'}).reset_index()
plt.bar(by_month.Months,by_month.Steps )
plt.show()

by_days = fitbit.groupby('Work_or_Weekend').agg({'Steps': 'median'}).reset_index()
plt.bar(by_days.Work_or_Weekend,by_days.Steps )
plt.ylabel('steps mean', fontsize= 12)
plt.xticks(by_days.Work_or_Weekend, ['Work days', 'Weekend'], fontsize= 12, rotation=45)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize= (18,9))

by_month = fitbit.groupby('Months').agg({'Steps': 'median'}).reset_index()
ax1.bar(by_month.Months,by_month.Steps )
for tick in ax1.get_xticklabels():
    tick.set_rotation(45)

by_days = fitbit.groupby('Work_or_Weekend').agg({'Steps': 'median'}).reset_index()
ax2.bar(by_days.Work_or_Weekend,by_days.Steps )
plt.ylabel('steps mean', fontsize= 12)
plt.xticks(by_days.Work_or_Weekend, ['Work days', 'Weekend'], fontsize= 12, rotation=45)
plt.show()
```

#### Plot the correlation matrix using seaborn.

```python
only_numerics = ['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
corr = titanic[only_numerics].corr()

sns.heatmap(corr)

plt.show()
```

```python
# Method 2 - seaborn
f, ax = plt.subplots(figsize=(11,9))

mask = np.triu(np.ones_like(corr, dtype=bool))
mask

# the color you want to use
cmap = sns.diverging_palette(220, 10, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap, annot=True, center=0, square=True, linewidths=0.5, cbar_kws={'shrink':0.5})
```

#### Use the most appropriate plot to display the summary statistics of `Age` depending on `Pclass`.

```python
mean_age = titanic.groupby('Pclass').agg({'Age': 'mean'}).reset_index()
std_age = titanic.groupby('Pclass').agg({'Age': 'std'}).reset_index()
min_age = titanic.groupby('Pclass').agg({'Age': 'min'}).reset_index()
max_age = titanic.groupby('Pclass').agg({'Age': 'max'}).reset_index()

plot_options , (chart_1,chart_2,chart_3,chart_4) = plt.subplots(nrows = 4, ncols = 1, figsize = (10,10), sharex=True)


# mean
chart_1.bar(mean_age['Pclass'], mean_age['Age'])
chart_1.title.set_text('mean')

# std
chart_2.bar(std_age['Pclass'], std_age['Age'])
chart_2.title.set_text('std')

# min
chart_3.bar(min_age['Pclass'], min_age['Age'])
chart_3.title.set_text('min')

# max
chart_4.bar(max_age['Pclass'], max_age['Age'])
chart_4.title.set_text('max')

plt.xticks([1,2,3])

plt.show()
```
